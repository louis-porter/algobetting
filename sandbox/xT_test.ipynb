```python
import pandas as pd

df = pd.read_csv(r"/Users/admin/Documents/dev/algobetting/infra/data/collectors/whoscored/events_202526.csv")
df['team'] = df['homeTeam'].where(df['h_a'] == 'h', df['awayTeam'])

df
```

```python
for col in df.columns:
    print(col)
```

```python
df_filtered = df[["matchId", "startDate", "minute", "second", "team", 'h_a', "x", "y", "type", "isTouch", "passAccurate", "passInaccurate", "EPV"]]
df_filtered = df_filtered[df_filtered["type"] == "Pass"]
df_filtered['startDate'] = pd.to_datetime(df_filtered['startDate'])

df_filtered[df_filtered['team'] == 'Arsenal']
```

```python
epv = df_filtered.groupby(["matchId", "team", 'startDate']).agg({
    "EPV": "sum"
}).reset_index()

epv[epv["matchId"] == 1903165]
```

```python
epv['season'] = '2025-26'
epv['division'] = 'Premier_League'

epv
```
